```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
f_13 = pd.read_csv('hofmeister_data_branch13.txt', sep=' ')
f_14 = pd.read_csv('hofmeister_data_branch14.txt', sep=' ')
print(f_14)
```

```python
branch_13 = np.array(f_13).T
branch_14 = np.array(f_14).T
dim_13 = np.shape(branch_13)
dim_14 = np.shape(branch_14)
only_SNP_13 = (branch_13 == 'RA').astype(int)
only_SNP_14 = (branch_14 == 'RA').astype(int)
SNPs_13=np.count_nonzero(only_SNP_13, axis=1)
SNPs_14=np.count_nonzero(only_SNP_14, axis=1)
SNPs_all=np.concatenate((SNPs_13,SNPs_14))
print(SNPs_13)
print(SNPs_14)
print(SNPs_all)
```

```python
x=['13.1', '13.2', '13.3', '13.5', '14.5', '14.4', '14.3', '14.2']
y=SNPs_all
fig, ax = plt.subplots(figsize=(10,7))

graph = ax.scatter(x,y)

points_to_annotate = [(1, 'Age:303'), (2, 'Age:284'), (3, 'Age:269'), (4, 'Age:193' ), (5, 'Age:200'), (6, 'Age:251'), (7, 'Age:299'), (8, 'Age:334')]

for point in points_to_annotate:
    ax.annotate(point[1], (x[point[0] - 1], y[point[0] - 1]), textcoords="offset points", xytext=(0, 5), ha='center')
ax.set_xlabel('Branch ID')
ax.set_ylabel('SNPs')
plt.show()
```

```python
x = [193, 200, 251, 269, 284, 299, 303, 334]
y = SNPs_all
fig, ax = plt.subplots(figsize=(10,7))

graph = ax.scatter(x,y)

points_to_annotate = [(1, 'Br. 13.5'), (2, 'Br. 14.5'), (3, 'Br. 14.4'), (4, 'Br. 13.3' ), (5, 'Br. 13.2'), (6, 'Br. 14.3'), (7, 'Br. 13.1'), (8, 'Br. 14.2')]

for point in points_to_annotate:
    ax.annotate(point[1], (x[point[0] - 1], y[point[0] - 1]), textcoords="offset points", xytext=(0, 5), ha='center')
ax.set_xlabel('Ages')
ax.set_ylabel('SNP number')
ax.set_title('Hofmeister et al, 2020')
plt.show()
```
